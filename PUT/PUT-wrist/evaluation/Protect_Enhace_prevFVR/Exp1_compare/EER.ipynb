```python
from bob.measure import eer
from bob.measure import eer_threshold
import numpy as np
root = '../../../models/'
FVRs=['img','wld','mc','rlt']
scenarios = ['normal','stolen']

def get_models(FVR,the_scenario):
    if the_scenario=='normal':
        scenario=''
    else:
        scenario='-'+the_scenario
    models = [
              'AE/alpha_1/emb_100/verify/results/'+FVR+'-AE-BHsh-100'+scenario,
              'Baseline/results/'+FVR+'-BHsh-100'+scenario,
              'Baseline/results/'+FVR
             ]
    return models

for scenario in scenarios:
    for FVR in FVRs:
        models = get_models(FVR,scenario)
        for model in models:
            print(scenario,FVR,model)
            try:
                scores_eval = root+FVR+'/'+model+'/baseline/wrist-R_1/nonorm/scores-eval'
                scores_dev  = root+FVR+'/'+model+'/baseline/wrist-R_1/nonorm/scores-dev'

                positives = []
                negatives = []
                with open(scores_dev,'r') as f:
                    lines = f.read().split('\n')
                    for line in lines:
                        line_splits = line.split()
                        if len(line_splits)==4:
                            if line_splits[0]==line_splits[1]:
                                positives.append(float(line_splits[3]))
                            else:
                                negatives.append(float(line_splits[3]))

                threshold = eer_threshold(negatives, positives)


                positives = []
                negatives = []
                with open(scores_eval,'r') as f:
                    lines = f.read().split('\n')
                    for line in lines:
                        line_splits = line.split()
                        if len(line_splits)==4:
                            if line_splits[0]==line_splits[1]:
                                positives.append(float(line_splits[3]))
                            else:
                                negatives.append(float(line_splits[3]))

                #print(threshold, eer_threshold(negatives, positives))
                #fmr,fnmr=fmr_fnmr(negatives, positives, threshold)
                eer_,fmr,fnmr=eer(negatives, positives,also_farfrr=True)
                print("FMR: {}, FNMR: {}, EER: {}".format(round(fmr*100,1),round(fnmr*100,1),round(eer_*100,1)))

            except:
                continue
```
